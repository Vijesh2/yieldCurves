```python
import ipywidgets as widgets
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
# Load your data
data = pd.read_excel("yieldData.xlsx", sheet_name="2. fwd curve")
data_cleaned = data.iloc[4:].copy()
data_cleaned.columns = data.iloc[2]
data_cleaned = data_cleaned.dropna(how="all")
data_cleaned = data_cleaned.set_index("years:")
```

```python
df_melted = data_cleaned.reset_index().melt(id_vars='years:', var_name='Term', value_name='Yield')
df_reformatted = df_melted.copy()
df_reformatted = df_reformatted.rename(columns={"years:": "curveDate"})

df_reformatted['curveDate'] = df_reformatted['curveDate'].astype(str)
unique_dates = df_reformatted['curveDate'].unique()

#df_reformatted = df_reformatted.set_index("years:")
```

```python
import plotly.express as px
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML

# Create the date picker
date_picker = widgets.DatePicker(description='Pick a Date', layout=widgets.Layout(width='30%'))

# Variable to hold the selected dates
dateList = []

# Function to handle date selection
def handle_date_change(change):
    if change['new'] and (change['new'] not in dateList):
        dateList.append(change['new'])
        refresh_display()

# Refresh the display to show the updated dateList
def refresh_display():
    clear_output(wait=True)

    # Create date labels and buttons, and pack them into individual HBoxes
    date_items = []
    for date in dateList:
        btn = widgets.Button(description='❌', layout=widgets.Layout(width='30px'))
        btn.date = date
        btn.on_click(remove_date)
        date_label = widgets.Label(str(date))
        date_items.append(widgets.HBox([date_label, btn]))

    # Pack all date HBoxes into a single VBox
    date_vbox = widgets.VBox(date_items)

    # Display the date picker on the left and date_vbox on the right inside an HBox
    display(widgets.HBox([date_picker, date_vbox]))

# Function to remove date from dateList
def remove_date(button):
    dateList.remove(button.date)
    refresh_display()

# Attach the function to the date picker
date_picker.observe(handle_date_change, names='value')

date_list_output = widgets.Output()


# Placeholder for the chart
chart_placeholder = widgets.Output()

# Create the top HBox
top_hbox = widgets.HBox([date_picker, date_list_output])

display(widgets.HBox([date_picker, date_list_output]))
```

```python
import pandas as pd
import plotly.express as px

def plot_selected_dates(df, selectedDates):
    # Filter the dataframe based on the selectedDates
    df_filtered = df[df['curveDate'].isin(selectedDates)]

    # Plot the chart
    fig = px.line(df_filtered, x='Term', y='Yield', color='curveDate', title="Yield Curve over Selected Dates")
    fig.show()

dateList_str = [str(date) for date in dateList]

plot_selected_dates(df_reformatted, dateList_str)
```
